# Fundamentos de Bases de Datos 


##Fundamentos de bases de datos y para ciencia de datos

En general, un sistema de bases de datos es una colección de datos interrelacionados y un conjunto de programas para acceder a ellos. Estas colección de datos, normalmente contiene información con diferentes niveles de relevancia para empresa. El objetivo principal del sistema es proporcionar una forma de almacenar y recuperar los datos de manera práctica y eficiente.

Los sistemas de bases de datos están diseñados para gestionar grandes cantidades de información. La gestión de datos implica la definición de estructuras para almacenar la información como la provisión de mecanismos para la manipulación de la información. Además los sistemas de base de datos deben garantizar la confiabilidad y seguridad de la información almacenada.

##Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos

Si bien Data Warehousing” se refiere al proceso de recolectar y manipular datos provenientes de diversas fuentes, con el fin de obtener información valiosas dposteriorment. Se entiende  como un depósito de datos bajo una plataforma utilizada para recolectar y analizar datos provenientes de múltiples fuentes heterogéneas. 

Este tipo de plataformas reúnen diversas tecnologías y componentes que permiten explotar los datos. Ocupa un lugar central dentro de para el desarrollo de soluciones de ciencia de datos.

Además de permite almacenar una gran cantidad de datos, así como también su tratamiento y análisis. Uno de sus objetivos primordiales es transformar los datos brutos en informaciones útiles, y hacerlos disponibles y accesibles para los usuarios. Permite a sus usuarios apoyarse en datos históricos y actuales para tomar mejores decisiones.

En la ciencia de datos, la información pueden provenir de muy diversas fuentes y ser tipo estructurada o no estructurada. Resulta "fundamental" asegurar la calidad de los datos para poder estar en posibilidad de desarrollar analisis confiables y de calidad.
Hoy en día los "almacenes" más usados son los llamados "data lakes" que tienen la posibilidad de conjuntar todo tipo de datos para su explotación.
