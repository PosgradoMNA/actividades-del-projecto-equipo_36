# Fundamentos de Bases de Datos 


ETL: extract, transform, and load.
This technique presents problems when working with very large datasets, scalability is not as easy and this can introduce stability problems to the data warehouse pipeline. 

ELT: It flips on itself but basically it makes the repository where the information is placed to be processed or transformed within it. This is used mostly with raw text files and it allows to transform larger datasets without damaging format.

Relational databases are organized in a way that they can be linked to each other in some form. No relational databases usually have their information stored by categories rather than relations.

```python
ndf = df.copy()
ndf.dropna(inplace = True) # Este método elimina toda la fila donde se encuentre algún valor faltante, por lo tanto se pueden perder muchos datos
ndf
```

```python
ndf = df.copy()
ndf.dropna(inplace = True, axis = 1) # axis 1 is columns / axis 0 is rows. 
ndf                                  # Elimina la fila si existe al menos un dato perdido en la columna
```

```python
ndf = df.copy()
ndf.dropna(how = 'all', inplace = True) # Elimina el renglón sólo si es que se encuentra la información perdida en todos los registros
ndf
```

```python
ndf = df.copy()
ndf.dropna(thresh = 5, inplace = True) #Necesita que en la fila haya 5 registros con valores nulos, para no ser eliminado del dataframe
ndf
```

```python
ndf = df.copy()
wm = ndf.X15.mean()

ndf['X15'].fillna(value = wm,
                     inplace = True) # Reemplazar los valores faltantes de la columna X15 con la media o promedio de la columna
ndf
```

```python
ndf['X14'].fillna(value = ndf.X14.median(), #Sintaxis alterna para llenar una columna, en este caso se llena con la media de los registros u observaciones
                    inplace = True)
ndf
```

```python
mm  = ndf.X1.mode() #Imprime la moda de la columna seleccionada, si hay varios datos empatados, se pueden mostrar con mm[K]
mm
```

```python
ndf['X1'].fillna(value = mm[0], 
                    inplace = True) #Llenar faltantes X1 con la moda
ndf
```

```python
ndf.dropna(subset=['X1', 'X2'], inplace = True) # Para remover datos de columnas en particular
ndf
```
