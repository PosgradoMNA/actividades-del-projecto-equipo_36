```python
Variance = np.cov([cdf.X6,cdf.Y])
Variance # Covarianza entre X6 y Y el resultado
```

```python
import seaborn as sns
```

```python
p = sns.regplot(data = cdf,  
                x = cdf.X6, 
                y = cdf.Y
               ) # Plot para graficar las razones de cambio lineales
p
```

```python
from sklearn.decomposition import PCA
pcs = PCA(n_components=2)

pcs.fit(cdf[['X6', 'Y']])

pcsSummary = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                          'Proportion of variance': pcs.explained_variance_ratio_,
                          'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)#column 
                          }
                          ) #proporciones de varianza utilizando análisis de componentes principales
pcsSummary
```

```python
pcsSummary = pcsSummary.transpose() # Transposición
pcsSummary.round(2)
```

```python
pcsSummary.columns = ['PC1', 'PC2'] # Resumen de los dos componentes escogidos
pcsSummary = pcsSummary.round(4)
pcsSummary
```

Con el análisis anterior podemos concluir que manteniendo estos dos componentes, uno de ellos atribuirá el 89% de la varianza y con el segundo componente el 11% restante. 


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
PC_components = np.arange(pcs.n_components_) + 1
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pcs.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1, 
                 y = np.cumsum(pcs.explained_variance_ratio_), # Este código toma en cuenta los componentes seleccionados para explicar la variación, 
                 color = 'black',                              # En caso de seleccionar todas las variables, se puede mostrar una gráfica esencialmente igual a un pareto
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Scree Plot')
plt.xlabel('N-th Principal Component')
plt.ylabel('Variance Explained')
plt.ylim(0, 1)
plt.show()
```
